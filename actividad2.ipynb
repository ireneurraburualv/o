```python
import requests
from lxml import html
import pandas as pd
```

# Análisis tienda Divina Martina

Queremos obtener todos los productos de la tienda Divina Martina en función de las distintas secciones con sus respectivos precios. Además de cada sección obtendremos también cual es el nombre del producto más barato con su precio también.

```python
#Creamos una lista con todas las secciones
secciones=["tops","pantalon","monos","vestidos-2","falda","chaquetas","calzado"]
datos_productos = []
for seccion in secciones:
    url = "https://www.divinamartina.com/collections/"+seccion #Recorremos cada seccion
    response = requests.get(url)
    if response.status_code == 200:
        root = html.fromstring(response.text)

        # Seleccionamos con xpath el elemento que contiene la descripción del producto
        articulos = root.xpath('//div[@class="ProductItem__Info ProductItem__Info--left"]')
        prenda_mas_barata = None #Guardamos cual es la prenda mas barata
        precio_mas_barato = float('inf')  # Inicializamos con infinito para garantizar que se actualice

        for articulo in articulos:
            nombre_producto = articulo.xpath('.//h2[@class="ProductItem__Title Heading"]')
            precio_producto = articulo.xpath('.//div[@class="ProductItem__PriceList  Heading"]')

            if len(nombre_producto) > 0:
                nombre = nombre_producto[0].text_content().strip()

                # Modificar para eliminar el símbolo del euro y convertir a número
                precio_texto = float(precio_producto[0].text_content().strip().replace("€", "").replace(",", "").replace(".", ""))

                # Actualizar la prenda más barata
                if precio_texto < precio_mas_barato:
                    prenda_mas_barata = nombre
                    precio_mas_barato = precio_texto

                datos_productos.append({"Nombre": nombre, "Precio": precio_texto, "Seccion": seccion})

        # Imprimir la prenda más barata de la sección actual
        if prenda_mas_barata is not None:
            print(f"Prenda más barata en {seccion}: {prenda_mas_barata} - €{precio_mas_barato:.2f}")

# Crear DataFrame
df_productos = pd.DataFrame(datos_productos)

# Mostrar el DataFrame
print(df_productos)
```
